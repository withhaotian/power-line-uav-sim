```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import ScalarFormatter


# Load data
data = pd.read_csv("drone_costs.csv")

# Create figure with improved layout
plt.figure(figsize=(10, 8))
plt.rcParams['font.size'] = 20
plt.rcParams['font.family'] = 'Arial'

# Define colors and styles
colors = ['black', 'dimgray', 'gray', 'darkgray']  # Tableau palette
colors = [str(0.1 + 0.4 * i/4) for i in range(4)]
line_styles = ['-', '-', '-', '-']
markers = ['o', 's', 'D', '^']
marker_size = 10

# Plot each drone's cost curve
for i, drone in enumerate(['Drone_0', 'Drone_1', 'Drone_2', 'Drone_3']):
    plt.plot(data['Iteration'], 
             data[drone], 
             label=f'UAV #{i+1}',
             color=colors[i],
             linestyle=line_styles[i],
             linewidth=2,
             marker=markers[i],
             markevery=9,
             markersize=marker_size)

# Customize axes and labels
plt.xlabel('Number of Iterations')
plt.ylabel('Cost of UAV')
# plt.xlim(0, len(data))
plt.ylim(0, data[['Drone_0', 'Drone_1', 'Drone_2', 'Drone_3']].max().max()*1.1)

# Improve grid and frame
plt.grid(True, linestyle='-', alpha=0.8, linewidth=1.5)
# Add legend
plt.legend()

# Set black border for all axes spines
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(1.5)

# Set y-axis to scientific notation with 1eX format
formatter = ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((0, 0))
ax.yaxis.set_major_formatter(formatter)

# Adjust layout and save
plt.tight_layout()
plt.savefig('drone_cost_convergence.pdf', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import ScalarFormatter


# Load data
data = pd.read_csv("base_station_utilities.csv")

# Create figure with improved layout
plt.figure(figsize=(10, 8))
plt.rcParams['font.size'] = 20
plt.rcParams['font.family'] = 'Arial'

# Define colors and styles
colors = [str(0.1 + 0.4 * i/8) for i in range(9)]  # Tableau palette
line_styles = ['-', '--']
markers = ['o', 's', 'D', '^', 'v', '<', '>', 'x', '*']
marker_size = 10

base_stations = [f'BS_{i}' for i in range(9)]

# Plot each drone's cost curve
for i, drone in enumerate(base_stations):
    plt.plot(data['Iteration'], 
             data[drone], 
             label=f'BS #{i+1}',
             color=colors[i],
             linestyle=line_styles[i%2],
             linewidth=2,
             marker=markers[i],
             markevery=9,
             markersize=marker_size)

# Customize axes and labels
plt.xlabel('Number of Iterations')
plt.ylabel('Utility of Base Station ($)')
# plt.xlim(0, len(data))
plt.ylim(0, data[base_stations].max().max()*1.1)

# Improve grid and frame
plt.grid(True, linestyle='-', alpha=0.8, linewidth=1.5)
# Add legend
plt.legend(ncol=3)

# Set black border for all axes spines
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(1.5)

# Set y-axis to scientific notation with 1eX format
formatter = ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((0, 0))
ax.yaxis.set_major_formatter(formatter)

# Adjust layout and save
plt.tight_layout()
plt.savefig('base_station_utilities_convergence.pdf', dpi=300)
plt.show()
```
