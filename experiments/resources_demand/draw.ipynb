```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import ScalarFormatter


# Load data
df = pd.read_csv("drone_costs.csv")
df.columns = df.columns.str.strip()

# Set parameters for the grouped bar chart
bar_width = 0.2  # Slightly reduce bar width for better separation
index = np.arange(len(df))

# Create figure with improved layout
plt.figure(figsize=(12, 8))
plt.rcParams['font.size'] = 24
plt.rcParams['font.family'] = 'Arial'

# Define colors and styles
# colors = ['#525252', '#969696', '#cccccc', '#f7f7f7']
colors = [str(1 - 0.8 * i/4) for i in range(4)]

# Plot bars with borders (new order: Random, Distance, Price, Ours)
plt.bar(index, df['random'], bar_width, label='Random', color=colors[0], edgecolor='black', linewidth=2, zorder=3)
plt.bar(index + bar_width + 0.02, df['distance'], bar_width, label='Distance', color=colors[1], edgecolor='black', linewidth=2, zorder=3)
plt.bar(index + 2 * bar_width + 0.02 * 2, df['price'], bar_width, label='Price', color=colors[2], edgecolor='black', linewidth=2, zorder=3)
plt.bar(index + 3 * bar_width + 0.02 * 3, df['ours'], bar_width, label='Ours', color=colors[3], edgecolor='black', linewidth=2, zorder=3)


# Customize axes and labels
plt.xlabel('Demand of Resources (Units)')
plt.ylabel('Averaged Cost of UAVs')
# plt.xlim(0, len(data))
plt.ylim(0, df[['random', 'distance', 'price', 'ours']].max().max()*1.5)
plt.xticks(index + 1.5*bar_width, df['resource'])

# Improve grid and frame
plt.grid(True, linestyle='-', alpha=0.8, linewidth=1.5)
# Add legend
plt.legend()

# Set black border for all axes spines
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(1.5)

# Set y-axis to scientific notation with 1eX format
formatter = ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((0, 0))
ax.yaxis.set_major_formatter(formatter)

# Adjust layout and save
plt.tight_layout()
plt.savefig('cost_vs_resources.pdf', dpi=300)
plt.show()
```
